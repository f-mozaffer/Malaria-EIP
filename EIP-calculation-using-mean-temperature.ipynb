```python
#Calculation of EIP using mean temperature (Month : May )for the site of Anusuya (2000 m)

from __future__ import division
import math
import matplotlib.pyplot as plt
import numpy as np
    

date=[]
temperature=[]

high_counter=[]
low_counter=[]
count_next_day=[]  

mean_daily_temperature=[]
mean_EIP=[]
DTR_temp=[]
std_EIP=[]

#initial values for daily temperature at an hour interval
count_low=0
count_high=0
date_counter=0
initial_counter=19052014003401

Vector_lifespan=56 #in days

#Briere equation 
def briere(x,c,To,Tm):
    if((x<To) or (x>Tm)):
        return 0.0
    else:
        return (c*x*(x-To)*math.sqrt(Tm-x))

# Input for the unknown values of parameter used in Briere equation    
def briere_input(temp):
    return(briere(temp,0.000116  ,15.624 ,34.921  ))

#Reading temperature input file    
Anusuya_temperature_value=open("Anusuya2014tempfile.txt","r")
temp_values_Anusuya=Anusuya_temperature_value.readlines()

#To extract date and temperature
for j in temp_values_Anusuya:
    temp_date=(j.strip('\n').split(' ')[0])
    temp_temperature=(j.strip('\n').split(' ')[1])
    temperature.append(temp_temperature)
    date.append(temp_date)


#To set a counter for a day at an hour interval
for k in range(len(date)):
   
    date_counter=int(date[k])
  
    if(date_counter==initial_counter):
    
        low_counter.append(initial_counter)
        last_counter=initial_counter+230000
        high_counter.append(last_counter)
        initial_counter=initial_counter+1000000000000



#To get mean daily temperature per day
for i in range(len(low_counter)):
    Daily_temp=[]
    for l in range(len(date)):
        if((int(date[l])>=low_counter[i]) and (int(date[l])<=high_counter[i])):
            Daily_temp.append(float(temperature[l]))
            
   
    mean_daily_temperature.append(np.mean(Daily_temp))
    DTR_temp.append(max(Daily_temp)-min(Daily_temp))

#EIP calculation    
for u in range(len(mean_daily_temperature)):
    pdr1=briere_input(round(mean_daily_temperature[u],3))
    if(pdr1==0):
        mean_EIP.append(np.nan)
        std_EIP.append(np.nan)
    else:
        eip_val=1/pdr1
        if(eip_val>90):
            mean_EIP.append(90)
            std_EIP.append(np.nan)
        else:
            mean_EIP.append(eip_val)
            std_EIP.append(eip_val)

#mean EIP for month of May            
EIP_using_mean_temp=np.nanmean(mean_EIP)
if(EIP_using_mean_temp>Vector_lifespan):
    EIP_using_mean_temp=np.nan

#mean standard deviation for month of May  
STD_using_mean_temp=np.nanstd(std_EIP)

```
